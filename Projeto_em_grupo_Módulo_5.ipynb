```python
#importando bibliotecas necessárias para o projeto
import pandas as pd
import numpy as np
import zipfile
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pandas as pd


# Caminho para o arquivo ZIP contendo os arquivos CSV
caminho_arquivo_zip = 'archive.zip'

# Lista dos nomes dos arquivos CSV das 10 criptomoedas selecionadas
nome_arquivo_csv = ['coin_Bitcoin.csv', 'coin_Ethereum.csv', 'coin_Monero.csv', 'coin_Stellar.csv', 'coin_NEM.csv', 'coin_XRP.csv', 'coin_Iota.csv', 'coin_Dogecoin.csv', 'coin_Tether.csv', 'coin_Litecoin.csv']

# Criar um DataFrame vazio para armazenar os dados
dados_criptomoedas = pd.DataFrame()

# Abrir o arquivo ZIP
with zipfile.ZipFile(caminho_arquivo_zip, 'r') as zipf:
    for arquivo in nome_arquivo_csv:
        with zipf.open(arquivo) as f:
            # Ler o arquivo CSV e combinar com os dados existentes
            df_temp = pd.read_csv(f)
            dados_criptomoedas = pd.concat([dados_criptomoedas, df_temp])

```

```python
dados_criptomoedas.fillna(dados_criptomoedas.mean(numeric_only=True), inplace=True)
```

```python
dados_criptomoedas.drop_duplicates(inplace=True)
```

```python
dados_criptomoedas['Date'] = pd.to_datetime(dados_criptomoedas['Date'])
```

```python
import matplotlib.pyplot as ml
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
import re


principal = pd.read_csv('criptomoedas.csv')
df = pd.DataFrame(principal)
df['Date'] = pd.to_datetime(df['Date'])

df

dfcor=df.pivot_table(columns='Name',index='Date',values='Close', aggfunc='mean',).reindex(pd.date_range(start='2020-09-01',end='2021-06-30'),method='nearest')
#dfcor
corrprincipal = dfcor.corr(method = 'pearson')

#corrprincipal
ml.figure(figsize=(8,6), dpi = 150)
vmin = 0
vmax = 1
annot = True
cmap = sns.light_palette("seagreen", as_cmap=True)

hm = sns.heatmap(data=corrprincipal, vmin=vmin,vmax=vmax,annot=annot, cmap=cmap)
hm.patch.set_facecolor('none')
ml.show()
```
